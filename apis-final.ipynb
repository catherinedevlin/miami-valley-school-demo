# [Studio Ghibli](https://ghibliapi.herokuapp.com/#)

```python
base_url = 'https://ghibliapi.herokuapp.com'
```

```python
resp = requests.get(base_url + '/people')
```

```python
print(resp.text)
```

```python
rawdata = resp.json()
```

```python
rawdata
```

```python
rawdata[2]['hair_color']
```

```python
rawdata[0].keys()
```

```python
import collections
hairs = collections.defaultdict(int)
```

```python
hairs['Pink'] = hairs['Pink'] + 1
```

```python
hairs
```

```python
grey = 0
for person in rawdata:
    if person['hair_color'] == 'Grey':
        grey += 1
print(grey)
```

```python
import collections
hairs = collections.defaultdict(int)
for person in rawdata:
    hairs[person['hair_color']] += 1
print(hairs)
```

```python
resp = requests.get(base_url + '/films')
for film in resp.json():
    print(film['title'])
```
